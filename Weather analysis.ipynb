```python
import pandas as pd 
import seaborn as sns
import matplotlib.pyplot as plt
import plotly.express as px
```

```python
data = pd.read_csv("weather.csv")
```

```python
data.head()
```

```python
data.info()
```

```python
data['Formatted Date'] = pd.to_datetime(data['Formatted Date'])
```

```python
plt.figure(figsize=(12, 6))
sns.lineplot(x='Formatted Date', y='Temperature (C)', data=data)
plt.title('Temperature Over Time')
plt.xlabel('Date')
plt.ylabel('Temperature (C)')
plt.xticks(rotation=45)
plt.show()
```

```python
numeric_columns = data.select_dtypes(include=[float, int])
```

```python
corr_matrix = numeric_columns.corr()
```

```python
plt.figure(figsize=(10, 8))
sns.heatmap(corr_matrix, annot=True, cmap='coolwarm', linewidths=0.5)
plt.title('Correlation Heatmap')
plt.show()
```

```python
plt.figure(figsize=(8, 6))
sns.histplot(data['Temperature (C)'], kde=True, bins=20)
plt.title('Temperature Distribution')
plt.xlabel('Temperature (C)')
plt.ylabel('Frequency')
plt.show()
```

```python
plt.figure(figsize=(8, 6))
sns.scatterplot(x='Temperature (C)', y='Apparent Temperature (C)', data=data)
plt.title('Scatter Plot: Temperature vs Apparent Temperature')
plt.xlabel('Temperature (C)')
plt.ylabel('Apparent Temperature (C)')
plt.show()
```

```python
plt.figure(figsize=(8, 6))
sns.scatterplot(x='Humidity', y='Wind Speed (km/h)', data=data)
plt.title('Scatter Plot: Humidity vs Wind Speed')
plt.xlabel('Humidity')
plt.ylabel('Wind Speed (km/h)')
plt.show()
```
